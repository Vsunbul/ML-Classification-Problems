```python
# import pandas and matplotlib libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read data
data = pd.read_excel("Worldbank GDP Data.xlsx", index_col = 0)
```

```python
data = data.fillna(0)/1000000000
```

```python
data.head()
```

```python
# Let's create a list of names containing countries.
countries = ["Germany", "France", "United Kingdom", "Italy", "Netherlands"]
```

```python
# let's see the countries data
data[countries].head()
```

```python
#increase the resolution of the figure
plt.figure(dpi = 150 )

# Create countries figure
plt.plot(data[countries], marker = "o", markevery = 5);

# Let's add information to the axes and title
plt.title("GDP Change from 1960 to 2019")
plt.ylabel("GDP (in billions of US $)")
plt.xlabel("Calendar Years")


plt.grid(visible = True, axis = "x", linestyle = "--")

for country in countries:
    plt.text(2019, data.loc[2018, country], int(data.loc[2018, country]), fontsize = 6)

# Add legend 
plt.legend(countries, ncol = 2)
plt.show()
```
